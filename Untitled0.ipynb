```python
import pandas as pd
```

```python
df=pd.read_csv("/content/data_ted_talks.csv")
```

```python
# prompt: linear regression graph

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Assuming 'views' and 'comments' are the columns you want to use for regression
X = df[['views']].values.reshape(-1, 1)
y = df['comments']

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the values
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Views')
plt.ylabel('Comments')
plt.title('Linear Regression of Views vs Comments')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example DataFrame with 'views' and 'comments'
data = {
    'views': [100, 200, 300, 400, 500, 600],
    'comments': [10, 20, 30, 40, 50, 60]
}
df = pd.DataFrame(data)

# Assuming 'views' and 'comments' are the columns you want to use for regression
X = df[['views']].values.reshape(-1, 1)
y = df['comments']

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict the values
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Views')
plt.ylabel('Comments')
plt.title('Linear Regression of Views vs Comments')
plt.show()
```
